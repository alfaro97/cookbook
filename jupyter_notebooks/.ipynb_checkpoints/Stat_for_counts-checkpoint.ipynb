```python
import matplotlib.pyplot as plt
import pandas
import os
import re
import pymongo
import json
from flask import Flask, jsonify, json
from flask_pymongo import PyMongo
from numpy import nan

DBS_NAME = os.getenv("DBS_NAME")
MONGO_URI = os.getenv("MONGODB_URI")

app = Flask(__name__)
app.config["DBS_NAME"] = "cookbook"
app.config["MONGO_URI"] = "mongodb://localhost:27017/cookbook"
mongo = PyMongo(app)

RECIPES_DATA_PATH = 'recipes.json'
cuisine_list = mongo.db.cuisine.find()

# READ JSON FILE
def read_json(file_path):
    results = []
    with open(RECIPES_DATA_PATH) as recipes_file:
        for recipes in recipes_file:
            try:
                recipe = json.loads(recipes)
                results.append(recipe)
            except ValueError:
                pass
        return results

# SEARCH JSON FOR TEXT
def is_text_in_field(token, recipe_content):
    token = token.lower()
    recipe_content = ''.join(recipe_content).lower()
    match = re.search(token, recipe_content)
    if match:
        return True
    return False

results = read_json(RECIPES_DATA_PATH)


# TOP 10 CUISINE
recipesDataFrame = pandas.DataFrame()
recipesDataFrame['cuisine'] = [recipe['cuisine'] for recipe in results]
for c in cuisine_list:
    recipesDataFrame[c['cuisine_name']] = recipesDataFrame['cuisine'].apply(lambda recipe: is_text_in_field(c['cuisine_name'], recipe))

recipesDataFrame['Category Counts'] = recipesDataFrame.groupby(['cuisine'])[c['cuisine_name']].transform('count')
# print(recipesDataFrame)


recipesCountsCuisDataFrame = pandas.DataFrame()
recipesCountsCuisDataFrame['Recipe Name'] = [recipe['name'] for recipe in results]
recipesCountsCuisDataFrame['Recipe Cuisine'] = [recipe['cuisine'] for recipe in results]
recipesCountsCuisDataFrame['Cuisine Counts'] = recipesCountsCuisDataFrame.groupby(['Recipe Cuisine'])['Recipe Cuisine']\
.transform('Cuisine Counts').value_counts()


# recipesCountsCuisDataFrame['Recipe Cuisine'].value_counts().reset_index(name='Cuisine Counts').rename(columns={'index': ['Recipe Cuisine']})

print(recipesCountsCuisDataFrame)

```
